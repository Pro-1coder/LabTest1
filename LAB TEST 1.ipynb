```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
aqi_readings = np.random.randint(200, 450, 1440)

# Noisy Data
def Noise(aqi_readings):
    jump = np.random.randint(1,50)
    for i in range(0, len(aqi_readings), jump):
        no = jump = np.random.randint(150,600)
        aqi_readings[i] = no
        jump = np.random.randint(40,50)

# aqi_readings[60:140:5] = 7
# aqi_readings[770:840:5] = 800
```

```python
from scipy import signal

aqi_readings = signal.filter
```

```python

def hourly(aqi_readings):
    Hourly_Aqi = [];
    for i in range(0, len(aqi_readings), 60):
        total = aqi_readings[i:i+60].mean()
        Hourly_Aqi.append(total)
    
    return Hourly_Aqi

Hourly_Aqi = hourly(aqi_readings)
len(Hourly_Aqi)
```

```python
plt.hist(aqi_readings, color='blue', label="AQI WITH NOISE")
plt.legend()
```

```python
plt.hist(Hourly_Aqi, color='blue', label="Hourly AQI")
plt.legend()
```

```python
def peak_interval(aqi_readings,threshold):
    threshold_Aqi = []
    time = []
    for i in range(0, len(aqi_readings)):
        total = aqi_readings[i:i+15].sum()
        if total >= threshold*15:
            threshold_Aqi.append(aqi_readings[i:i+15])
            for j in range(i, i+15):
                time.append(j)
        return threshold_Aqi, time
    
# plt.scatter(np.arange(0, 1440), aqi_readings)
threshold_Aqi, time = peak_interval(aqi_readings, 300)
threshold_Aqi = np.array(threshold_Aqi)
time = np.array(time)

time
```
